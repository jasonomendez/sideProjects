```python

import numpy as np 
import matplotlib.pyplot as plt 
```

```python
class statsData: 
    def __init__(self, data): 
        self.data = data 
 
    def compute_sample_range(self):
        return np.max(self.data) - np.min(self.data) 


    def compute_sample_variance(self): 
        return np.var(self.data, ddof=1) 


    def compute_sample_standard_deviation(self): 
        return np.sqrt(self.compute_sample_variance())  


    def sample_median(self): 
        return np.median(self.data) 


    def sample_mean(self): 
        return np.mean(self.data)  

    def mode(self):
        return np.argmax(np.bincount(self.data)) 


    def sample_quartiles(self): 
            # Quartiles as defined by medians 
            quartiles = np.percentile(self.data, [25, 50, 75])
            return quartiles
    

    def interQuartileRange(self): 
        Q1, Q2, Q3 = self.sample_quartiles()
        return Q3 - Q1


    def outliers(self): 
        Q1, _, Q3 = self.sample_quartiles() 
        IQR = self.interQuartileRange() 
        lower_bound = Q1 - 1.5*IQR 
        upper_bound = Q3 + 1.5*IQR  
        outliers = [] 
        for i in self.data: 
            if i < lower_bound or i > upper_bound: 
                outliers.append(i) 
        return outliers
        
    def compute_statistics(self): 
        range = float(self.compute_sample_range())
        variance = float(self.compute_sample_variance())
        std_dev = float(self.compute_sample_standard_deviation())
        median = float(self.sample_median())
        mean = float(self.sample_mean() )
        quartiles = tuple(map(float, self.sample_quartiles()))
        IQR = float(self.interQuartileRange()) 
        outliers = list(map(float, self.outliers()))
        return {
            "sample range": range, 
            "sample variance": variance, 
            "Sample Standard Deviation": std_dev, 
            "sample median": median,
            "sample mean": mean, 
            "quartiles": quartiles,  
            "interquartile range": IQR, 
            "outliers": outliers
        }
```

```python
def plot_boxplot(data): 
    plt.boxplot(data)
    plt.title('Boxplot of Data')
    plt.show()

def plot_scatterplot(data): 
    plt.scatter(data[:,0], data[:,1])
    plt.title('Scatterplot of Data')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()


def plot_lineplot(data): 
    plt.plot(data)
    plt.title('Lineplot of Data')
    plt.xlabel('Index')
    plt.ylabel('Value')
    plt.show()


def plot_histogram(data): 
    plt.hist(data, bins = 10, alpha = 0.5, color = 'blue', edgecolor = 'black')
    plt.title('Histogram of Data')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.show()
 

```

```python
data = np.array([0, 0, 0, 0, 0.1, 0.1, 0.1, 0.1, 0.2, 0.2, 0.3, 0.3, 0.3, 0.4, 0.5, 0.7, 0.8, 1.0, 1.5, 2.7, 2.8, 3.5, 4.0, 8.9, 9.2, 11.7, 21.0])

# Compute and display all statistics 
data_inference = statsData(data)
data_statistics = data_inference.compute_statistics()
print(data_statistics)

# Compute and display the mode
plot_boxplot(data)
```
